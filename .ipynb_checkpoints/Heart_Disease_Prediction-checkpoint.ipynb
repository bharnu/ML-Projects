```python
for col in ['male', 'age', 'education', 'currentSmoker', 'BPMeds',
            'prevalentStroke', 'prevalentHyp', 'diabetes']:

    params_mle_no_chd[col] = None
    params_mle_chd[col] = None
```

```python
for col in ['male', 'currentSmoker', 'BPMeds',
            'prevalentStroke', 'prevalentHyp', 'diabetes']:

    p_hat_no_chd = len(no_chd_training_data[no_chd_training_data[col] == 1])/len(no_chd_training_data)
    q_hat_no_chd = len(no_chd_training_data[no_chd_training_data[col] == 0])/len(no_chd_training_data)

    p_hat_chd = len(chd_training_data[chd_training_data[col] == 1])/len(chd_training_data)
    q_hat_chd = len(chd_training_data[chd_training_data[col] == 0])/len(chd_training_data)

    params_mle_no_chd[col] = {0:q_hat_no_chd,1:p_hat_no_chd}
    params_mle_chd[col] = {0:q_hat_chd,1:p_hat_chd}
```

```python
for age in data['age'].unique():

  if len(no_chd_training_data[no_chd_training_data['age'] == age]) == 0:
    params_mle_no_chd['age'][age] = 1/(len(no_chd_training_data) + len(data['age'].unique()))
  else:
    params_mle_no_chd['age'][age] = len(no_chd_training_data[no_chd_training_data['age'] == age])/len(no_chd_training_data)

  if len(chd_training_data[chd_training_data['age'] == age]) == 0:
    params_mle_chd['age'][age] = 1/(len(chd_training_data) + len(data['age'].unique()))
  else:
    params_mle_chd['age'][age] = len(chd_training_data[chd_training_data['age'] == age])/len(chd_training_data)
```

```python
import scipy.stats as s
```

```python
def compute_no_chd_likelihood(all_feature_values):

  joint_probability_no_chd = 1

  for feature_value_idx in range(0,4):

    joint_probability_no_chd = joint_probability_no_chd * params_mle_no_chd[data.columns[feature_value_idx]][all_feature_values[feature_value_idx]]

  joint_probability_no_chd = joint_probability_no_chd * s.poisson.pmf(all_feature_values[4],params_mle_no_chd[data.columns[4]])

  for feature_value_idx in range(5,9):

    joint_probability_no_chd = joint_probability_no_chd * params_mle_no_chd[data.columns[feature_value_idx]][all_feature_values[feature_value_idx]]

  remaining_feature_values = all_feature_values[9:]

  joint_probability_no_chd = joint_probability_no_chd * s.multivariate_normal.pdf(remaining_feature_values,
                                                                                  params_mle_no_chd["mean_vector"],
                                                                                  params_mle_no_chd["covariance_matrix"])
  
  return joint_probability_no_chd
```

```python
def compute_chd_likelihood(all_feature_values):

  joint_probability_chd = 1

  for feature_value_idx in range(0,4):

    joint_probability_chd = joint_probability_chd * params_mle_chd[data.columns[feature_value_idx]][all_feature_values[feature_value_idx]]

  joint_probability_chd = joint_probability_chd * s.poisson.pmf(all_feature_values[4],params_mle_chd[data.columns[4]])

  for feature_value_idx in range(5,9):

    joint_probability_chd = joint_probability_chd * params_mle_chd[data.columns[feature_value_idx]][all_feature_values[feature_value_idx]]

  remaining_feature_values = all_feature_values[9:]

  joint_probability_chd = joint_probability_chd * s.multivariate_normal.pdf(remaining_feature_values,
                                                                                  params_mle_chd["mean_vector"],
                                                                                  params_mle_chd["covariance_matrix"])
  
  return joint_probability_chd
```

```python
cv_data = np.array(cv_data)
```

```python
gt_cv_labels = cv_data[:,15]
```

```python
gt_cv_labels
```

```python
cv_data = cv_data[:,0:15]
```

```python
cv_data.shape
```

```python
predicted_cv_labels = list()

for i in range(0,cv_data.shape[0]):

  P_X_on_CHD = compute_chd_likelihood(cv_data[i,:])

  P_X_on_NoCHD = compute_no_chd_likelihood(cv_data[i,:])

  P_CHD_on_X = P_X_on_CHD/(P_X_on_CHD + P_X_on_NoCHD)

  if P_CHD_on_X > 0.5:

    predicted_cv_labels.append(1)

  else:

    predicted_cv_labels.append(0)
```

```python
predicted_cv_labels = np.array(predicted_cv_labels)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_true=gt_cv_labels,y_pred=predicted_cv_labels))
```
